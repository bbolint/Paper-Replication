```python
epinions_data = pd.read_excel('soc-sign-epinions.xlsx', header = 3)
epinions_data
```

```python
# Read in the Wikipedia dataset

wikipedia_data = pd.read_excel('wikipedia_raw.xlsx')
```

```python
# Get elections as separate objects (actually part of the same object, but separable)

election_index = wikipedia_data[wikipedia_data.iloc[:,0] == "E"].index

election_index = list(election_index)

votes = np.split(wikipedia_data, election_index)
```

```python
# Count the number of votes casted in each election

voter_count = list()

for i in range(len(votes)):
    voter_count.append(len(votes[i][votes[i].iloc[:,0] == "V"].index))
    
voter_count = voter_count[1:]
```

```python
# Create a list to be added as the "to" column

nominee_ids = list(wikipedia_data[wikipedia_data.iloc[:,0] == "U"].iloc[:,1])

nominee_col = []

for i in range(len(nominee_ids)):
    nominee_col += [nominee_ids[i]] * voter_count[i]
```

```python
# Add the newly created "to" column to the dataset containing details of specific votes only

elections = wikipedia_data[wikipedia_data.iloc[:,0] == "V"]
elections.columns = ['row content', 'vote(1:support, 0:neutral, -1:oppose)', 'voter_id', 'time', 'voter_username']

elections['nominee_id'] = nominee_col

elections = elections[['voter_id', 'nominee_id', 'vote(1:support, 0:neutral, -1:oppose)', 'time']]
elections
```
